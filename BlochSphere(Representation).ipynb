```python
# Otras librerías y la librería QUTIP entera
import numpy as np
import matplotlib.pyplot as plt
from qutip import *
```

```python
# Graficación de Qubits
num_qubits = 1
q = qutip.basis(num_qubits, 0)
state = rand_ket(2**num_qubits)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
sphere = Bloch(axes=ax)
sphere.add_states(state)
sphere.show()
```
