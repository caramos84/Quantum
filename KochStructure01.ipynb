```python
# Fractales: Estructuras de Koch
# Carlos Ramos + ChatGPT o1
# 30/032025

import matplotlib.pyplot as plt
import numpy as np
```

```python
def koch_curve(p1, p2, depth, graph):
    if depth == 0:
        plt.plot([p1[0], p2[0]], [p1[1], p2[1]], 'k')
        graph.append((p1, p2))  # Guardamos la conexi√≥n entre nodos
        return

    # Calculamos los puntos intermedios
    p3 = (2*p1[0] + p2[0]) / 3, (2*p1[1] + p2[1]) / 3
    p5 = (p1[0] + 2*p2[0]) / 3, (p1[1] + 2*p2[1]) / 3

    # Rotamos el segmento central
    angle = np.pi / 3
    dx, dy = p5[0] - p3[0], p5[1] - p3[1]
    p4 = (p3[0] + dx * np.cos(angle) - dy * np.sin(angle),
          p3[1] + dx * np.sin(angle) + dy * np.cos(angle))

    # Llamadas recursivas
    koch_curve(p1, p3, depth - 1, graph)
    koch_curve(p3, p4, depth - 1, graph)
    koch_curve(p4, p5, depth - 1, graph)
    koch_curve(p5, p2, depth - 1, graph)
```

```python
def generate_koch_hypergraph(iterations):
    plt.figure(figsize=(10, 5))
    graph = []  # Lista para almacenar los enlaces

    # Puntos iniciales de la curva de Koch
    p1, p2 = (0, 0), (1, 0)
    koch_curve(p1, p2, iterations, graph)

    plt.axis("equal")
    plt.axis("off")
    plt.show()

    return graph  # Devolvemos la estructura de hipergrafo

# Generar la curva de Koch con estructura de hipergrafo
hypergraph = generate_koch_hypergraph(5)
```
