```python
# Simulación de atractor de Lorenz
# Carlos Ramos + ChatGPT 4o
# 21/03/2025

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.integrate import solve_ivp
```

```python
# Definimos el sistema de ecuaciones de Lorenz
def lorenz(t, state, sigma=10, beta=8/3, rho=100):
    x, y, z = state
    dx_dt = sigma * (y - x)
    dy_dt = x * (rho - z) - y
    dz_dt = x * y - beta * z
    return [dx_dt, dy_dt, dz_dt]
```

```python
# Condiciones iniciales y tiempo de simulación
initial_state = [1.0, 1.0, 1.0]
t_span = (0, 50)
t_eval = np.linspace(t_span[0], t_span[1], 10000)

# Resolver el sistema de ecuaciones
solution = solve_ivp(lorenz, t_span, initial_state, t_eval=t_eval, method='RK45')
x, y, z = solution.y
```

```python
# Crear la figura y graficar el atractor de Lorenz
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5, color='blue')
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.show()
```
